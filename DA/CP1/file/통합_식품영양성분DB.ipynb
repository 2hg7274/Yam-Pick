### 1차 전처리

```python
# NaN 제거
df.columns = df.iloc[2]
df.drop(index=[0,1,2,3], inplace=True)

# 불필요한 column 제거
df.drop(columns=df.iloc[:, :5].columns, inplace=True)
df.drop(columns=df.iloc[:, 1:4].columns, inplace=True)
df.drop(columns=['총내용량(g)', '총내용량(mL)', '성분표출처', '발행기관'], inplace=True)

df.reset_index(drop=True, inplace=True)
```

```python
# 표기 되지 않은 영양성분 '-' -> 0으로 변경
df.iloc[:, 5:].replace('-', 0, inplace=True)

# '1g 미만' -> 0으로 변경
df.iloc[:, 5:].replace('1g 미만', 0, inplace=True)
```

```python
df['에이코사디에노산(20:2(n-6))(g)'] = df['도코사헥사에노산(22:6(n-3))(g)'] + df['에이코사디에노산(20:2(n-6))(g)']
df.drop(columns='도코사헥사에노산(22:6(n-3))(g)', inplace=True)
df.rename(columns={'에이코사디에노산(20:2(n-6))(g)': 'EPA+DHA(g)'}, inplace=True)

df['레티놀(㎍)'] = df['베타카로틴(㎍)'] + df['레티놀(㎍)']
df.drop(columns='베타카로틴(㎍)', inplace=True)
df.rename(columns={'레티놀(㎍)': '비타민 A(㎍)'}, inplace=True)

df['토코트리에놀(㎎)'] = df['토코페롤(㎎)'] + df['토코트리에놀(㎎)']
df.drop(columns='토코페롤(㎎)', inplace=True)
df.rename(columns={'토코트리에놀(㎎)': '비타민 E(㎎)'}, inplace=True)
```
