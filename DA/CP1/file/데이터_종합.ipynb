```python
import pandas as pd
import os
```

```python
sum_df = pd.DataFrame()

year_range = [x for x in range(2008, 2021)]

for year in year_range:
    data = pd.read_excel(f'/Users/simcheol-u/Desktop/여자/식품군별_평균_섭취량_연령층별(여자)_{year}.xls')
    data['연도'] = year
    sum_df = pd.concat([sum_df, data], ignore_index=False)
```

```python
delete_list = sum_df.columns[sum_df.columns.str.contains('표준')]
```

```python
sum_df.drop(columns=delete_list, inplace=True)
sum_df['성별'] = '여성'
```

```python
sum_df.to_csv('/Users/simcheol-u/Desktop/식품군별_평균_섭취량_연령층별(여성)_종합', index=False)
```
